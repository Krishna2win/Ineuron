```python
2. What are the three different types of Boolean operators?

==>

AND: 
    The and operator returns True if both operands (values or expressions) on its left and right are True. 
    Otherwise, it returns False. In other words, for the and operator to yield True, all the operands must be True.
    
OR: 
    The or operator returns True if at least one of the operands on its left or right is True. 
    It returns False only if both operands are False.
    
NOT: 
    The not operator is a unary operator that negates the Boolean value of its operand. 
    If the operand is True, not will return False, and if the operand is False, not will return True.
```

```python
5. What are the six comparison operators?

==>

The six comparison operators in Python are used to compare two values or expressions and return a Boolean
result (True or False) based on the comparison. Here are the six comparison operators:

Equal to: ==
Checks if two values or expressions are equal.

Not equal to: !=
Checks if two values or expressions are not equal.

Greater than: >
Checks if the value or expression on the left is greater than the one on the right.

Less than: <
Checks if the value or expression on the left is less than the one on the right.

Greater than or equal to: >=
Checks if the value or expression on the left is greater than or equal to the one on the right.

Less than or equal to: <=
Checks if the value or expression on the left is less than or equal to the one on the right.
```

```python
6. How do you tell the difference between the equal to and assignment operators?Describe a
condition and when you would use one.

==>

Equal to Operator ==:

The equal to operator is a comparison operator used to check whether two values or expressions are equal.
It returns a Boolean value (True or False) based on the comparison.
Example: x == y checks if the value of x is equal to the value of y.

Assignment Operator =:
    
The assignment operator is used to assign a value to a variable.
It takes the value on its right and stores it in the variable on its left.
Example: x = 10 assigns the value 10 to the variable x.
    
You would use the equal to operator (==) when you need to compare values or expressions to check for equality.
This is often used in conditional statements (if, elif, else) or when evaluating conditions in loops and functions.
    
You would use the assignment operator (=) when you want to assign a value to a variable. 
This is common when initializing variables or updating their values during the program's execution.

```

```python
11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

==>

range(10):

This expression generates a sequence of numbers from 0 to 9 (10 numbers in total).
The range() function by default starts from 0 and stops just before the specified end value (exclusive). In this case,
the end value is 10, so the sequence includes numbers 0, 1, 2, ..., 9.
It uses the default step value of 1, which means the numbers in the sequence are incremented by 1 at each step.

range(0, 10):

This expression also generates a sequence of numbers from 0 to 9 (10 numbers in total).
The range() function starts from the specified start value, which is 0, and stops just before the specified end value (exclusive), which is 10.
Like the previous case, it uses the default step value of 1.

range(0, 10, 1):

This expression is equivalent to the previous two examples.
It explicitly specifies the start value as 0, the end value as 10, and the step value as 1.
Since the step value is 1, it increments the numbers in the sequence by 1 at each step.
```

```python
12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent
program that prints the numbers 1 to 10 using a while loop.

==>

# Using a for loop
for i in range(1, 11):
    print(i)

    
# Using a while loop
i = 1
while i <= 10:
    print(i)
    i += 1
```
