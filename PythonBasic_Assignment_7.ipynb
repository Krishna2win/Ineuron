```python
7. The findall() method returns a string list or a list of string tuples. What causes it to return one of the two options?

==>

The findall() method in Python is re module returns either a list of strings or a list of string tuples based 
on the presence of capturing groups in the regular expression.

List of Strings:
    If the regular expression used in findall() does not contain any capturing groups (i.e., no parentheses in the pattern), 
    then the method returns a list of strings. Each element in the list represents a full match of the pattern in the input string.

List of String Tuples:
    If the regular expression contains one or more capturing groups (i.e., one or more sets of parentheses in the pattern), 
    then the findall() method returns a list of string tuples. 
    Each tuple corresponds to a full match of the entire regular expression, and each element of the tuple represents 
    the text matched by a specific capturing group.
```

```python
9. In regular expressions, what does the character stand for?

==>

In regular expressions, the term "character" typically refers to any individual symbol or character from the input text 
that the regular expression is attempting to match or find. It can be a literal character, a metacharacter, 
or a special character with a specific meaning, depending on the context and the regular expression pattern used.

Here are some common types of characters in regular expressions:

Literal Characters: 
    These are characters that match themselves exactly in the input text. 
    For example, the regular expression pattern r'apple' will match the word "apple" wherever it appears in the input text.

Metacharacters: 
    These are characters with special meanings in regular expressions. For instance:

. (dot): Matches any single character except a newline.
*: Matches zero or more occurrences of the preceding character or group.
+: Matches one or more occurrences of the preceding character or group.
?: Matches zero or one occurrence of the preceding character or group.
^: Matches the start of the string.
$: Matches the end of the string.
Character Classes: These are special sequences that represent a group of characters. For example:

\d: Matches any digit (equivalent to [0-9]).
\w: Matches any word character (letters, digits, or underscore, equivalent to [a-zA-Z0-9_]).
\s: Matches any whitespace character (spaces, tabs, newlines, etc.).
\D, \W, \S: Negations of \d, \w, \s, respectively.
    
Escaped Characters: 
    Some characters have special meanings in regular expressions, 
    but if you want to match them literally, you can escape them using a backslash \.
    For example, to match a literal dot . or a literal parenthesis ( or ), you would use \. or \( or \).
```

```python
10.In regular expressions, what is the difference between the + and * characters?

==>

In regular expressions, the + and * characters are quantifiers that control the number of occurrences of the preceding 
character or group. They are used to match patterns with varying repetition of characters.

* (Asterisk):

    The * quantifier matches zero or more occurrences of the preceding character or group. 
    It means the character or group can appear any number of times, including zero.
    For example, the pattern a* will match strings with any number of consecutive 'a' characters, 
    including no 'a' at all. It will match "", "a", "aa", "aaa", and so on.

+ (Plus):

    The + quantifier matches one or more occurrences of the preceding character or group. 
    It means the character or group must appear at least once for a match to occur.
    For example, the pattern a+ will match strings with one or more consecutive 'a' characters. 
    It will match "a", "aa", "aaa", and so on, but not "" (empty string).
```

```python
11. What is the difference between {4} and {4,5} in regular expression?

==>

In regular expressions, {4} and {4,5} are both quantifiers used to specify the exact number of occurrences of the
preceding character or group. However, they differ in the number of occurrences they match:

{4}:

    The {4} quantifier specifies exactly four occurrences of the preceding character or group.
    For example, the pattern a{4} will match exactly four consecutive 'a' characters.

{4,5}:

    The {4,5} quantifier specifies a range of occurrences, from a minimum of four to a maximum of five occurrences 
    of the preceding character or group.
    For example, the pattern a{4,5} will match strings with either four or five consecutive 'a' characters.
```

```python
12. What do you mean by the \d, \w, and \s shorthand character classes signify in regular expressions?

==>

In regular expressions, shorthand character classes are predefined character classes represented by certain escape sequences. 
They provide a convenient way to match common groups of characters without having to specify each character individually. 
Shorthand character classes are widely used to match digits, word characters, and whitespace characters.

Here are the meanings of the shorthand character classes:

\d:

    \d is a shorthand for "digit." It matches any digit from 0 to 9.
    For example, the pattern \d will match any single digit in the input text.

\w:

    \w is a shorthand for "word character." It matches any alphanumeric character (letters A-Z, a-z, and digits 0-9)
    as well as the underscore _.
    For example, the pattern \w+ will match one or more word characters in the input text.

\s:

    \s is a shorthand for "whitespace character." It matches any whitespace character, including spaces, tabs, newlines, etc.
    For example, the pattern \s+ will match one or more whitespace characters in the input text.
```

```python
13. What do means by \D, \W, and \S shorthand character classes signify in regular expressions?

==>

In regular expressions, the shorthand character classes \D, \W, and \S are negations of their respective counterparts
\d, \w, and \s. They represent predefined character classes that match characters that are NOT digits, word characters, 
and whitespace characters, respectively.

Here is what each shorthand character class signifies:

\D:

    \D is the negation of \d. It matches any character that is NOT a digit (i.e., any character that is not in the range 0-9).
    For example, the pattern \D+ will match one or more non-digit characters in the input text.

\W:

    \W is the negation of \w. It matches any character that is NOT a word character 
    (i.e., any character that is not a letter, digit, or underscore).
    For example, the pattern \W+ will match one or more non-word characters in the input text.

\S:

    \S is the negation of \s. It matches any character that is NOT a whitespace character
    (i.e., any character that is not a space, tab, newline, etc.).
    For example, the pattern \S+ will match one or more non-whitespace characters in the input text.
```

```python
14. What is the difference between .*? and .*?

==>

In regular expressions, the difference between .*? and .* lies in their behavior when used for pattern matching. 
Both are used to match any sequence of characters, but they operate with different matching modes:

.*? (Non-Greedy/Lazy Quantifier):

    The .*? is a non-greedy (lazy) quantifier. It matches as few characters as possible while still allowing the rest of the pattern to match.
    In a non-greedy mode, the quantifier tries to find the smallest possible match for the subsequent pattern elements.
    For example, in the pattern a.*?b, if the input text is "aabab", it will match "aab" as the smallest match, 
    whereas in a greedy mode, it would have matched the whole string "aabab".

.* (Greedy Quantifier):

    The .* is a greedy quantifier. It matches as many characters as possible while still allowing the rest 
    of the pattern to match.
    In a greedy mode, the quantifier tries to find the largest possible match for the subsequent pattern elements.
    For example, in the pattern a.*b, if the input text is "aabab", it will match the entire string "aabab" 
    because it is the largest match that still satisfies the rest of the pattern.
```

```python
16. What is the procedure for making a normal expression in regax case insensitive?

==>

To make a regular expression case-insensitive in Python, you can use the re.IGNORECASE or re.
I flag (both are equivalent) as the second argument to the relevant regex function.

There are multiple regex functions in Pythons re module, such as re.search(), re.findall(), re.sub(), etc. 
All of these functions can take an optional flags parameter as the second argument.

Heres the procedure for making a regular expression case-insensitive:

Use re.IGNORECASE or re.I as the flags parameter when calling a regex function.
 for example :
        
import re

text = "Find matches for 'apple' and 'Orange' using case-insensitive regex."

pattern = r'apple|orange'

matches = re.findall(pattern, text, flags=re.IGNORECASE)

print("Matches:", matches)
```

```python
 20. How would you write a regex that match a number with comma for every three digits? It must
match the given following:
'42'
'1,234'
'6,368,745'
but not the following:
'12,34,567'(which has only two digits between the commas)
'1234' (which lacks commas)

==>


import re

pattern = r'^\d{1,3}(,\d{3})*$'

test_strings = ['42', '1,234', '6,368,745', '12,34,567', '1234']

for text in test_strings:
    if re.match(pattern, text):
        print(f"'{text}' matches.")
    else:
        print(f"'{text}' does not match.")

```

```python
21. How would you write a regex that matches the full name of someone whose last name is
Watanabe? You can assume that the first name that comes before it will always be one word that
begins with a capital letter. The regex must match the following:
'Haruto Watanabe'
'Alice Watanabe'
'RoboCop Watanabe'
but not the following:
'haruto Watanabe' (where the first name is not capitalized)
'Mr. Watanabe' (where the preceding word has a nonletter character)
'Watanabe' (which has no first name)
'Haruto watanabe' (where Watanabe is not capitalized)

==>

import re

pattern = r'^[A-Z][a-zA-Z]*\sWatanabe$'

test_names = ['Haruto Watanabe', 'Alice Watanabe', 'RoboCop Watanabe', 'haruto Watanabe', 'Mr. Watanabe', 
              'Watanabe', 'Haruto watanabe']

for name in test_names:
    if re.match(pattern, name):
        print(f"'{name}' matches.")
    else:
        print(f"'{name}' does not match.")
```

```python
22. How would you write a regex that matches a sentence where the first word is either Alice, Bob,
or Carol; the second word is either eats, pets, or throws; the third word is apples, cats, or baseballs;
and the sentence ends with a period? This regex should be case-insensitive. It must match the
following:
'Alice eats apples.'
'Bob pets cats.'
'Carol throws baseballs.'
'Alice throws Apples.'
'BOB EATS CATS.'
but not the following:
'RoboCop eats apples.'
'ALICE THROWS FOOTBALLS.'
'Carol eats 7 cats.'

==>

import re

pattern = r'^(Alice|Bob|Carol)\s(eats|pets|throws)\s(apples|cats|baseballs)\.$'

test_sentences = [
    'Alice eats apples.',
    'Bob pets cats.',
    'Carol throws baseballs.',
    'Alice throws Apples.',
    'BOB EATS CATS.',
    'RoboCop eats apples.',
    'ALICE THROWS FOOTBALLS.',
    'Carol eats 7 cats.'
]

for sentence in test_sentences:
    if re.match(pattern, sentence, re.IGNORECASE):
        print(f"'{sentence}' matches.")
    else:
        print(f"'{sentence}' does not match.")
```
